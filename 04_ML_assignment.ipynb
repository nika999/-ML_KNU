```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
training = pd.read_csv("pml-training.csv")
testing = pd.read_csv("pml-testing.csv")
training.head()
```

```python
testing.head()
```

## Clean Data

```python
training.drop(labels=training.columns[0:6],axis=1,inplace=True)
testing.drop(labels=testing.columns[0:6],axis=1,inplace=True)
```

```python
training = training.dropna(1)
testing = testing.dropna(1)
```

```python
training.head()
```

```python
print("number of rows : ", training.shape[0])
print("number of columns : ", training.shape[1])
```

```python
testing.head()
```

```python
print("number of rows : ", testing.shape[0])
print("number of columns : ", testing.shape[1])
```

## Next, we get rid of some columns that do not contribute much to the accelerometer measurements.

```python
training = training[training.columns.drop(list(training.filter(regex='timestamp')))]
training = training[training.columns.drop(list(training.filter(regex='window')))]
training = training[training.columns.drop(list(training.filter(regex='user_name')))]
training = pd.get_dummies(training, columns = ['classe'])
```

```python
training['classe_A'] = training['classe_A'].replace([1],0)
training['classe_B'] = training['classe_B'].replace([1],1)
training['classe_C'] = training['classe_C'].replace([1],2)
training['classe_D'] = training['classe_D'].replace([1],3)
training['classe_E'] = training['classe_E'].replace([1],4)
training['classe'] = training['classe_A'] +  training['classe_B']+  training['classe_C']+  training['classe_D']+  training['classe_E']
training = training.drop(['classe_A'], axis = 1)
training = training.drop(['classe_B'], axis = 1)
training = training.drop(['classe_C'], axis = 1)
training = training.drop(['classe_D'], axis = 1)
training = training.drop(['classe_E'], axis = 1)
training.head()
```

```python
testing = testing[testing.columns.drop(list(testing.filter(regex='timestamp')))]
testing = testing[testing.columns.drop(list(testing.filter(regex='window')))]
testing = testing[testing.columns.drop(list(testing.filter(regex='user_name')))]
testing.head()
```

```python
dataset = training
#y = dataset['classe_A' ,'classe_B' ,'classe_C' ,'classe_D' ,'classe_E' ]
y = dataset['classe']
X = dataset.drop(['classe'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

## We fit a predictive model for activity recognition using Random Forest algorithm because it automatically selects important variables and is robust to correlated covariates & outliers in general. 

```python
scaler= StandardScaler()

X_scaled=scaler.fit_transform(X_train)
X_scaled=pd.DataFrame(X_scaled, columns = X.columns)

X_valid=scaler.transform(X_test)
X_valid=pd.DataFrame(X_test, columns = X.columns)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model_1=XGBClassifier(random_state=0,eval_metric='mlogloss')
model_1.fit(X_scaled,y_train)
predictions = pd.DataFrame(model_1.predict(X_valid))
accuracy_score1= accuracy_score(y_test,predictions)*100

formatted_float = "{:.2f}%".format(accuracy_score1)
print("Model accuracy:" ,formatted_float)
```

## Predicting for Test Data Set

```python
testing = testing.drop(['problem_id'], axis = 1)
X_test_scaled=scaler.transform(testing)
X_test_scaled=pd.DataFrame(X_test_scaled, columns = testing.columns)
preds = model_1.predict(X_test_scaled)
```

```python
result = pd.DataFrame(preds, columns = ['Predict'])
```

```python
result['Predict'] = result['Predict'].replace([0],'A')
result['Predict'] = result['Predict'].replace([1],'B')
result['Predict'] = result['Predict'].replace([2],'C')
result['Predict'] = result['Predict'].replace([3],'D')
result['Predict'] = result['Predict'].replace([4],'E')

result['real'] = ['B','A','B','A','A','C','D','B','A',
                        'A','B','C','B','A','E','E','A','B','B','B']
```

```python
accuracy_score2= accuracy_score(result['real'],result['Predict'])*100
formatted_float = "{:.2f}%".format(accuracy_score2)
print("Model accuracy:" ,formatted_float)
```

```python
frames = [result, testing]
result2 = pd.concat(frames, axis=1)
```
